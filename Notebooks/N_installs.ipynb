```python
import json
import pandas as pd
import os
import glob
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data_path = "/content/drive/MyDrive/VPN Deprecated/data"
output_path = "/content/drive/MyDrive/VPN Deprecated/Output"
folders = [p for p in glob.glob(data_path + "/*") if os.path.isdir(p)]
```

```python
arr = []
for folder in folders:
  app = folder.split('/')[-1]
  file_json = f"{folder}/{app}_playstore.json"
  with open(file_json) as f:
    data = json.load(f)
  n_installs = data['realInstalls']
  summarized_installs = data['installs']
  arr.append([app, n_installs, summarized_installs])
```

```python
df = pd.DataFrame(arr, columns=['app', 'n_installs', 'summarized_installs'])
df.to_csv(f"{output_path}/n_installs.csv", index=False)
```
