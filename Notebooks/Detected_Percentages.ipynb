```python
import json
import pandas as pd
import glob, os
```

```python
out_path="/content/drive/MyDrive/VPN Deprecated/Output"
folders = [p for p in glob.glob(out_path + "/*") if os.path.isdir(p)]
```

```python
total = 0
total_servers = 0
arr = []

for folder in folders:
  app = folder.split('/')[-1]
  json_file = f"{folder}/total.json"

  with open(json_file , 'r') as f:
    data = json.load(f)
  total = data['total']
  vpns = data['vpn']

  arr.append([app, total, vpns])
```

```python
df = pd.DataFrame(arr, columns=['app', 'total', 'vpns'])
```

```python
total = df['total'].sum()
vpns = df['vpns'].sum()
df['percentage'] = (df['vpns'] / df['total']) * 100
```

```python
df.sort_values(by=['total'], ascending=False, inplace=True)
```

```python
df.to_csv(f"{out_path}/detected.csv", index=None)
```
