```python
import json
import pandas as pd
import glob, os
```

```python
path="/content/drive/MyDrive/VPN Deprecated/data"
out_path="/content/drive/MyDrive/VPN Deprecated/Output"
folders = [p for p in glob.glob(path + "/*") if os.path.isdir(p)]
```

```python
arr = []

for folder in folders:
  app = folder.split('/')[-1]
  csv_file = f"{folder}/{app}.csv"
  df_app = pd.read_csv(csv_file)
  total = len(df_app)

  arr.append([app, total]) # -1 cause of the header
```

```python
df = pd.DataFrame(arr, columns=['app', 'servers'], index=None)
```

```python
total = df['servers'].sum()
df['percentage'] = (df['servers'] / total) * 100
```

```python
df.sort_values(by=['percentage'], ascending=False, inplace=True)
df.info()
```

```python
df.to_csv(f"{out_path}/n_servers.csv", index=None)
```
