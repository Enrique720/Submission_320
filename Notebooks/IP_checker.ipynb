```python
# Description: This code is to verify that all applications have all their IPs scanned by censys. Comparison: csv vs json.
import json
import glob, os
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
path="/content/drive/MyDrive/VPN Deprecated/data"
out_path="/content/drive/MyDrive/VPN Deprecated/Recollect"
folders = [p for p in glob.glob(path + "/*") if os.path.isdir(p)]
```

```python
total = 0
for folder in folders:
  app = folder.split('/')[-1]
  csv = f"{folder}/{app}.csv"
  js_file = f"{folder}/{app}.json"
  ips = set()
  tot = 0
  try:
    with open(js_file, 'r') as f:
      #print(js_file)
      for line in f:
        data = json.loads(line)
        ip = data['result']['resource']['ip']
        ips.add(ip)

    f_csv = open(csv, 'r')
    outf = open(f"{out_path}/{app}.csv", 'w')
    for line in f_csv:
      ip = line.split(',')[0]
      if ip not in ips:
        outf.write(line)
        tot+=1
    outf.close()
    f_csv.close()
  except:
    continue

  print(f"app: {app} total: {tot}")
```
